####Risk score v/s login_frequency
**Conclusion**
- The analysis of risk scores against login frequency reveals that high-risk users predominantly have low login frequency, suggesting potentially anomalous or one-off behavior â€” a common pattern in fraudulent or malicious activity. On the other hand, users with consistent and frequent logins tend to exhibit lower to moderate risk levels, likely due to more predictable behavioral patterns

```python

# Create color mapping for 'anomaly'
colors = {'low_risk': 'green', 'moderate_risk': 'orange', 'high_risk': 'red'}

# Apply colors to each point based on anomaly
color_values = df['anomaly'].map(colors)

plt.figure(figsize=(10, 6))
plt.scatter(df['login_frequency'], df['risk_score'], c=color_values, alpha=0.5)

plt.title('Risk Score vs Login Frequency (Colored by Anomaly)')
plt.xlabel('Login Frequency')
plt.ylabel('Risk Score')
plt.grid(True)

# Create custom legend
handles = [plt.Line2D([0], [0], marker='o', color='w', label=label,
                      markersize=8, markerfacecolor=color)
           for label, color in colors.items()]
plt.legend(handles=handles, title="Anomaly Level")

plt.show()
```

##FEATURE ENGINEERING

####Creating a is_fraud column
- for clear classification of fraud i.e scam and phising
- categorising transaction type into fraud or non fraud

```python
fraud = ['scam','phising']
df['is_fraud'] = df['transaction_type'].isin(fraud).astype(int)
df[['transaction_type','is_fraud']].head(10)
```

####Categorizing Risk into low, moderate and high

```python
# Define risk buckets based on risk_score
def get_risk_bucket(score):
    if score < 30:
        return 'Low'
    elif 30 <= score < 70:
        return 'Medium'
    else:
        return 'High'

# Apply the function to create a new column
df['risk_bucket'] = df['risk_score'].apply(get_risk_bucket)

# Optional: check the distribution
print(df['risk_bucket'].value_counts())
```
